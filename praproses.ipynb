```python
# handling data frame
import pandas as pd
# retrieve files/pathnames matching a specified pattern
import glob
```

# Praproses Data

## Merge Data

```python
# path of the folder
path = r'C:\Users\Siti Nur Hasanah\penelitian\data covid'

# display the list of excel files
filenames = glob.glob(path + "\*.xlsx")

# initializing the empty data frame
data = pd.DataFrame()

# iterate excel file one by one inside the folder
for file in filenames:
    
    # combining multiple excel worksheets into single data frame
    df = pd.concat(pd.read_excel(file, sheet_name=None), ignore_index=True)
    
    # appending excel files one by one
    data = data.append(df, ignore_index=True)
    
# display the final excel sheet
data.head(3)
```

```python
#simpan data ke folder
data.to_excel('dataset_jun.xlsx', index=False)
```

## Cleaning Data

1. Hapus semua nilai yang kolom kabupaten/kota NaN
2. Hapus nilai yg bukan nama kabupaten/kota
3. Hapus kolom selain (terakhir update, kabupaten/kota, sembuh, meninggal, konfirmasi)

```python
# hapus kalo ada white space di header kolom
data.rename(mapper=str.strip, axis='columns')
# ubah header jadi huruf kecil semua 
data.columns = [x.lower().strip() for x in df.columns]
data
```

```python
# kolom apa saja yg ada di data frame
data.columns
```

```python
# copy data frame
data_copy = data.copy()
```

Proses berikutnya akan dilakukan di data copy

```python
'''
    Simpan cuma kolom yang diperlukan
'''
columns_to_keep = ['terakhir update', 'kabupaten/kota', 'konfirmasi (hari ini)', 
                   'sembuh (hari ini)', 'meninggal (hari ini)', 'fatality rate (hari ini)', 'konfirmasi (sebelum)']
data_copy = data_copy[columns_to_keep]
data_copy.head()
```

```python
# ubah format tanggal (tanpa waktu)
data_copy['terakhir update'] = pd.to_datetime(data_copy['terakhir update']).dt.date
data_copy.head(10)
```

```python
# cek nilai yg ada di kolom kabupaten/kota
data_copy['kabupaten/kota'].unique()
```

```python
# simpan cuma baris yg bukan nan
data_copy = data_copy[data_copy['kabupaten/kota'].notna()]
data_copy.shape
```

```python
# cek apakah masih ada nan di data set
print(data_copy.isna().sum())
```

```python
'''
    hapus yg bukan nama kabupaten/kota
'''
# Get names of indexes for which column kabupaten/kota has value AWAK BUAH KAPAL
idx = data_copy[data_copy['kabupaten/kota'] == 'AWAK BUAH KAPAL'].index

# Delete these row indexes from dataFrame
data_copy.drop(idx, inplace=True)
data_copy.shape
```

```python
# Get names of indexes for which column kabupaten/kota has value RS LAPANGAN INDRAPURA
idx = data_copy[data_copy['kabupaten/kota'] == 'RS LAPANGAN INDRAPURA'].index

# Delete these row indexes from dataFrame
data_copy.drop(idx, inplace=True)
data_copy.shape
```

```python
# simpan data bersih ke folder
data_copy.to_excel('data_clean.xlsx', index=False)
```

```python
#urutkan per kabupaten/kota per tanggal
data_copy = (data_copy.sort_values(['kabupaten/kota','terakhir update'], ascending=[True, True])
            .set_index('kabupaten/kota'))
data_copy
```

```python
# simpan data terurut ke folder
data_copy.to_excel('covid-june.xlsx', index=True)
```
